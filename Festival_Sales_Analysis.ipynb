**(EDA) Exploratory Data Analysis**

*Gender*

```python
df.columns
```

```python
plot = sns.countplot(x = 'Gender', data = df)

for bars in plot.containers:
    plot.bar_label(bars)
```

```python
df.columns
```

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gender, hue = 'Gender')
```

```python
plt.figure(figsize=(3,4))
sns.barplot(x='Gender', y='Amount', data=sales_gender)
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales Amount')
plt.tight_layout()
plt.show()
```

*From above Bar plots, we can see that most of the buyers are female with more spending power too*

**Age**

```python
df.columns
```

```python
plot = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
for bars in plot.containers:
    plot.bar_label(bars)
```

```python
# Total amount vs Age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

*From above we can see that most of the buyers are of age group b/w 26-35 years mostly females*

**State**

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
# Total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

*From above graphs we can see that most of the Orders & Sales are from U.P., Maharashtra and Karnataka respectively*

**Marital Status**

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_state, hue ="Gender")
```

*From above graph we can see that most of the buyers are married women and they have more purchasing power also*

**Occupation**

```python
sns.set(rc={'figure.figsize':(28,9)})
plot = sns.countplot(x = 'Occupation', data = df)
for bars in plot.containers:
    plot.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={'figure.figsize':(26,8)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
```

*From above graphs we can observe that most of the buyers are from IT, Healthcare and Aviation sector respectively*

**Product Category**

```python
sns.set(rc={'figure.figsize':(30,10)})
plot = sns.countplot(x = 'Product_Category', data = df)
for bars in plot.containers:
    plot.bar_label(bars)
```

```python
sales_state = df.groupby(["Product_Category"], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_state)
```

*From above graphs we can observe that the most sold product categories are Food, Clothing and Electronics respectively*

**Product ID**

```python
df.columns
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,6)})
sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_state)
```

```python
# (By Different command) top 10 most sold products
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby(['Product_ID'])['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

**Conclusive Insights**

*Married women of age group 26-35 years from U.P., Maharashtra and Karnataka working in IT, Healthcare and Aviation are mostly the customers to buy products from Food, Clothing and Electronics respectively.*
